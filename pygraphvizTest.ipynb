```python
# Libraries
import pygraphviz as pgv
from IPython.display import Image
from IPython.display import SVG
from IPython.core.display import HTML
```

```python
# Creates a graph
G = pgv.AGraph(strict=False,directed=True)
G.add_nodes_from(['a','b','c'], color='black',style='filled',fillcolor='white')
G.add_edge('b','c')
G.add_edge('a','a')
G.add_edge('a','b')
```

```python
# Draws a movement of the DFA
def movement(a,b):
    G.get_node(a).attr['fillcolor'] = 'white'
    G.get_node(b).attr['fillcolor'] = 'palegreen'
    
# Draws the complete graph
def drawGraph(n):
    filename = 'file' + str(n) + '.png'
    G.draw(filename, prog='dot')
    return Image(filename=filename)
```

```python
# Draws the image
drawGraph(1)
```

```python
G.get_node('a').attr['fillcolor'] = 'palegreen'
drawGraph(2)
```

```python
movement('a','b')
drawGraph(3)
```

```python
movement('b','c')
drawGraph(4)
```
